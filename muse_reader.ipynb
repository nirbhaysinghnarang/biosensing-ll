```python
import csv 
import pandas as pd

f_path = "muse_eeg_data_2024-09-19T15_49_45.291Z.csv"
data = pd.read_csv(f_path)  # Behold! The data emerges from its slumber.



grouped_data = data.groupby(['timestamp', 'electrode']).agg({
    'value': 'mean'  # Calculating the average of the 'value' column
}).reset_index()
print(grouped_data.head())
grouped_data_sorted = grouped_data.sort_values(by='timestamp')

# Pivot the table to get electrodes as columns and timestamps as rows
time_series_grouped = grouped_data_sorted.pivot(index='timestamp', columns='electrode', values='value')

print(time_series_grouped.head())  # Display the first few rows of the grouped time series data
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Assuming 'time_series_grouped' is your data frame ready for the spotlight
time_series_grouped.plot(kind='line', figsize=(10, 6))

plt.title('Electrode Readings Over Time')  # Giving your masterpiece a title
plt.xlabel('Timestamp')  # Ah, the x-axis, the timeline of our plot's story
plt.ylabel('Value')  # The y-axis, where the data points reach for the stars
plt.legend(title='Electrode')  # A legend, because every story needs a guide
plt.xticks(rotation=45)  # Twisting those timestamps so they don't bump into each other
plt.tight_layout()  # Ensuring everything fits nicely in the frame
plt.show()  # And voila, the grand reveal!

```
